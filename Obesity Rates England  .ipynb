```python
# This notebook is from following tutorial
# http://pythonforengineers.com/data-analysis-with-pandas/
# Analyse obesity in England
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Reads excel file
data = pd.ExcelFile("Obes-phys-acti-diet-eng-2014-tab.xls")
print(data.sheet_names)
```

```python
# We are making columns for our data frame
columns1 = ['year', 'total', 'males', 'females']
```

```python
# Now get parse data from excel file
data_gender = data.parse(u'7.1', skiprows=4, skipfooter=14, names=columns1)
print(data_gender)
```

```python
# Drops rows if any values are missing 
data_gender.dropna(inplace = True)
print(data_gender)
```

```python
# Currently the index is 1,2,3,.... What we need us year as index
data_gender.set_index('year', inplace=True)
print(data_gender)
```

```python
data_gender.plot()
plt.show()
```

```python
# Returns first n rows with head(n=5)
data_gender.head()
```

```python
data_age = data.parse('7.2', skiprows=4, skipfooter=14)
data_age.head()
```

```python
# Rename the unamed row to year
data_age.rename(columns={'Unnamed: 0': 'Year'}, inplace=True)
data_age.head()
```

```python
# Drop empty rows
data_age.dropna(inplace=True)
```

```python
# Plotting 
data_age.plot()
plt.show()
```

```python
# In the above graph the total is too huge and supress other graphs. 
# So we need to remove the total to get a better look 
data_age_minus_total = data_age.drop('Total', axis=1)
data_age_minus_total.plot()
plt.show()
```

```python
# In the above graph we can see that age groups 45-54 and 45-44 are under the highest risk for obesity
# Now we are going to compare two age groups
# Here we are selecting a age group
print(data_age['Under 16'])
data_age['Under 16'].plot(label="Under 16")
data_age['25-34'].plot(label="25-34")
plt.legend(loc="upper right")
plt.show()
```
